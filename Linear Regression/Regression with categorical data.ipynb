```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data=pd.read_csv('Dummies.csv')
```

```python
raw_data
```

```python
data=raw_data.copy()
```

```python
data['Attendance']=data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
data.describe()
```

```python
y=data['GPA']
x1=data[['SAT','Attendance']]
```

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],data['GPA'],c=data['Attendance'],cmap='RdYlGn_r')
yno=0.6439+0.0014*data['SAT']
yyes=0.8665+0.0014*data['SAT']
fig1=plt.plot(data['SAT'],yno,lw=2,c='orange',label='resgression line1')
fig2=plt.plot(data['SAT'],yyes,lw=2,c='green',label='resgression line2')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
new_data=pd.DataFrame({'const':1,'SAT':[1700,1670],'Attendance':[0,1]})
new_data=new_data[['const','SAT','Attendance']]
```

```python
new_data
```

```python
predictions=results.predict(new_data)
predictions
```

```python
predictionsdf=pd.DataFrame({'Predictions':predictions})
print(predictionsdf)
joined=new_data.join(predictionsdf)
joined.rename(index={0:'BOB',1:'ALICE'})
```
