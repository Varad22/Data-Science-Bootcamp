```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data= pd.read_csv('simple_linear_regression.csv')
```

```python
data
```

```python
data.describe()
```

```python
y=data['GPA']
x1=data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
x=sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat=0.0017*x1+.275
fig=plt.plot(x1,yhat,lw=4,c='orange',label='regression line')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```
